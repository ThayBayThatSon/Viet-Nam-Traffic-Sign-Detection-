# Bước 3: Huấn luyện Mô hình YOLOv8
Tiến hành huấn luyện với 50 epochs (vòng lặp). Quá trình này sẽ tốn khoảng 50 - 75 phút trên GPU T4.

```python
from ultralytics import YOLO

# Sử dụng mô hình YOLOv8n (nano)
model = YOLO('yolov8n.pt') 

# Bắt đầu huấn luyện
results = model.train(
    data='/content/datasets/yolo_dataset/data.yaml', 
    epochs=50,
    imgsz=640,
    batch=16,
    name='traffic_sign_model'
)
```
